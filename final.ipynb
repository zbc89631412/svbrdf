```python
def filename_sort(directory_name):
  file_name_list = []
  for file_name in os.listdir(directory_name):
    file_name_list.append(file_name)
  file_name_list.sort()
  return file_name_list
```

```python
file=docx.Document()

reference_directory_name = "/content/gdrive/MyDrive/final/reference_cropped"

features_directory_name=["/content/gdrive/MyDrive/final/model1_cropped","/content/gdrive/MyDrive/final/model2_cropped","/content/gdrive/MyDrive/final/model3_cropped"]
gt = filename_sort(reference_directory_name)
# model1 = filename_sort(features_directory_name[0])
# model2 = filename_sort(features_directory_name[1])
# model3 = filename_sort(features_directory_name[2])
print(gt)
model_count = 0
feature_count = 0
image_count = 0
rmse_file=docx.Document()
for model in features_directory_name:
   fake_filename_list = filename_sort(model)
   print(fake_filename_list)
   rmse_diff = []
   rmse_normal =[]
   rmse_spec = []
   rmse_roughness=[]
   table = file.add_table(rows=84, cols=4, style="Table Grid")
   row=0
   file.add_paragraph("new model")
   for fake_img_name,gt_img_name in zip(fake_filename_list,gt):
     fake_img = cv2.imread(model+"/"+fake_img_name,0)
     gt_img = cv2.imread(reference_directory_name+"/"+gt_img_name,0)
     rmse = rmsdiff(fake_img,gt_img)
     if gt_img_name.find("diffuse")!=-1:
       rmse_diff.append(rmse)
       #table.cell(row,0).text = str(rmse)
     if gt_img_name.find("normal")!=-1:
       rmse_normal.append(rmse)
       #table.cell(row,1).text=str(rmse)
     if gt_img_name.find("specular")!=-1:
       rmse_spec.append(rmse)
       #table.cell(row,2).text=str(rmse)
     if gt_img_name.find("roughness")!=-1:
       rmse_roughness.append(rmse)
       #table.cell(row,3).text=str(rmse)
   for i in range(len(rmse_diff)):
     table.cell(i,0).text =str(round(rmse_diff[i],4))
     table.cell(i,1).text = str(round(rmse_normal[i],4))
     table.cell(i,2).text = str(round(rmse_roughness[i],4))
     table.cell(i,3).text = str(round(rmse_spec[i],4))
  #  table.cell(len(rmse_diff),0)=str(round(np.mean(rmse_diff),4))
  #  table.cell(len(rmse_normal),1)=str(round(np.mean(rmse_normal),4))
  #  table.cell(len(rmse_roughness),2)=str(round(np.mean(rmse_roughness),4))
  #  table.cell(len(rmse_spec),3)=str(round(np.mean(rmse_spec),4))
file.save("/content/gdrive/MyDrive/final/rmse.docx")
   
    
   
    
    
```
